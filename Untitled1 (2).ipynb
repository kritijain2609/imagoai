## 1. Data Exploration and Preprocessing

```python
import pandas as pd
import numpy as np
import matplotlib.pyplot as plt
import seaborn as sns
from sklearn.preprocessing import StandardScaler
from sklearn.decomposition import PCA
from sklearn.model_selection import train_test_split
from sklearn.ensemble import RandomForestRegressor
from sklearn.metrics import mean_absolute_error, mean_squared_error, r2_score
```

```python
df = pd.read_csv("TASK-ML-INTERN.csv")  
print(df.head())  
print(df.info())  
```

```python
df.fillna(df.select_dtypes(include=["number"]).mean(), inplace=True)
```

```python
target_column = "vomitoxin_ppb"  
if target_column in df.columns:
    features = df.drop(columns=[target_column])
    target = df[target_column]
else:
    raise KeyError(f"Column '{target_column}' not found in DataFrame!")
    
features = features.select_dtypes(include=["number"])
    
# Standardize features
scaler = StandardScaler()
features_scaled = scaler.fit_transform(features)
```

## 2. Dimensionality Reduction

```python
pca = PCA(n_components=2)
features_pca = pca.fit_transform(features_scaled)

# Visualize PCA results
plt.scatter(features_pca[:, 0], features_pca[:, 1], c=target, cmap='viridis')
plt.colorbar(label="vomitoxin_ppb")
plt.xlabel("Principal Component 1")
plt.ylabel("Principal Component 2")
plt.title("PCA Visualization")
plt.show()
```

## 3. Model Training

```python
X_train, X_test, y_train, y_test = train_test_split(features_scaled, target, test_size=0.2, random_state=42)
model = RandomForestRegressor(n_estimators=100, random_state=42)
model.fit(X_train, y_train)
```

## 4. Model Evaluation

```python
y_pred = model.predict(X_test)
mae = mean_absolute_error(y_test, y_pred)
rmse = np.sqrt(mean_squared_error(y_test, y_pred))
r2 = r2_score(y_test, y_pred)

print(f"MAE: {mae}")
print(f"RMSE: {rmse}")
print(f"R² Score: {r2}")
```

```python
plt.scatter(y_test, y_pred, alpha=0.5)
plt.xlabel("Actual Values")
plt.ylabel("Predicted Values")
plt.title("Actual vs. Predicted")
plt.show()
```
